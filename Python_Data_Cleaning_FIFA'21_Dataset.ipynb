```python
fifa['Club'].unique
```

```python
def remove_whitespace(dataset):
 
    for i in dataset.columns:
        if dataset[i].dtypes == 'object':
            dataset[i] = dataset[i].astype(str).map(str.strip)
        else:
            pass

        
remove_whitespace(fifa)
fifa.head(10)
```

## 11. Contract

- ### This column has more than one information such as contract 'start year', 'end year' and also we can compute 'contract_length'.
- ### We will split this column into 3 columns i.e. Contract_start, Contract_end and Contract_length
- ### We will then categorize the Contract as 'Free', 'On Loan' and 'In Contract' by adding one more new column called as Contract_status.

```python
#Unique values in the column

fifa["Contract"].unique()
```

```python
# We will split this column value into 3 new columns i.e Start Year, End Year, Contract Length.
def extract_contract_info(contract):
    if contract=='Free' or 'On Loan' in contract:
        start_year=''
        end_year=''
        contract_length=0
    else:
        start_year,end_year=contract.split(' ~ ')
        start_year=int(start_year[:4])
        end_year=int(end_year[:4])
        contract_length=end_year-start_year
    return start_year,end_year,contract_length

new_cols=['Contract_start','Contract_end','Contract_length (Years)']
new_data=fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i,column=new_cols[i],value=new_data[i])
```

```python
fifa.head()
```

```python
# Categorize Contract Status

def contract_status(contract):
    if contract=='Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'In Contract'

fifa.insert(loc=fifa.columns.get_loc('Contract_length (Years)')+1,column='Contract_Status',value=fifa['Contract'].apply(contract_status))
fifa.head()
```

```python
fifa['Contract_Status'].unique()
```

## 12. Height

- ### In this coulmn, height values are populated with 2 units i.e. cm and inches.
- ### We will convert all the values to cm and also will remove the unit 'cm' and retain only the value.
- ### We will rename the Height column to Height(cm)

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

fifa['Height']=fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
fifa.rename(columns={'Height':'Height(cm)'},inplace=True)

fifa.head()
```

## 13. Weight

- ### In this coulmn, weight values are populated with 2 units i.e. kg and lbs.
- ### We will convert all the values to kg and also will remove the unit 'kg' and retain only the value.
- ### We will rename the Weight column to Weight(kg)

```python
fifa['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds=int(weight.strip("lbs"))
        return round(pounds/2.205)

fifa['Weight']=fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

## 17. Value
- ### The Value column contains market value of each player.
- ### Every Value is suffixed either with 'M' or 'K' and prefixed with '€' symbol.
- ### We will get rid of 'M' and 'K' by multiplying the values by 1000000 and 1000 correspondingly.
- ### We will also remove the '€' symbol.

```python
fifa['Value'].unique()
```

```python
def convert_value(value):
    if "M" in value:
        a=value.strip("M").strip("€")
        return round(float(a)*1000000)
    else:        
        a=value.strip("K").strip("€")
        return round(float(a)*1000)

fifa['Value']=fifa['Value'].apply(convert_value)
```

```python
fifa['Value'].unique()
```

## 18. Wages
- ### The column contains wage of each player.

```python
fifa['Wage'].unique()
```

```python
def convert_wage(wage):
    if "K" in wage:
        a=wage.strip("K").strip("€")
        return int(a)*1000
    else:
        a=wage.strip("€")
        return int(a)

fifa['Wage']=fifa['Wage'].apply(convert_wage)
```

## 20. Hits
- ### Some Values are suffixed with 'K'.
- ### We will get rid of 'K' by multiplying the values by 1000.

```python
fifa['Hits'].unique()
```

```python
def convert_hits(hits):
    if "K" in hits:
        a=hits.strip("K")
        return round(float(a)*1000)
    else:
        if len(hits) <= 0:
            return 0;
        else:
            hits=hits.strip()
            return round(float(hits))
        

fifa['Hits']=fifa['Hits'].apply(convert_hits)
```
